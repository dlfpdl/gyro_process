```python
import pandas as pd
```

```python
def max_value():        
    with open('gyro_record_20190626-153444.csv', 'r') as f:
        lines = f.readlines()[1:]
        for bucket in range(0, len(lines) - 49, 49):
            x = []
            y = []
            for idx in range(bucket, bucket + 49):
                line = lines[idx].split(',')
                x.append(float(line[4]))
                q = max(x)
                y.append(q)
            return y
    
def min_value():
    with open('gyro_record_20190626-153444.csv', 'r') as f:
        lines = f.readlines()[1:]
        for bucket in range(0, len(lines) - 49, 49):
            x = []
            y = []
            for idx in range(bucket, bucket + 49):
                line = lines[idx].split(',')
                x.append(float(line[4]))
                q = min(x)
                y.append(q)
            return y
```

```python
M = max_value()
m = min_value()
M_value = pd.DataFrame(M)
m_value = pd.DataFrame(m)
```

```python
def run_out_value():
    with open('gyro_record_20190626-153444.csv', 'r') as f:
        lines = f.readlines()[1:]
        for bucket in range(0, len(lines) - 49, 49):
            for idx in range(bucket, bucket + 49):
                mean = (M_value - m_value) / len(lines)
                return mean
```

```python
run_out_value()
```
